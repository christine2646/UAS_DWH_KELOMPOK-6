```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(
    'P_Data_Extract_From_Health_Nutrition_and_Population_Statistics.xlsx',
    sheet_name='Data'
)

df.head()
```

### Mengubah Missing Value

Nilai `..` pada dataset diubah menjadi `NaN` agar dapat diproses sebagai missing value.

```python
df = df.replace('..', np.nan)
```

### Mengubah kolom tahun ke numeric

```python
years = [
    '2020 [YR2020]',
    '2021 [YR2021]',
    '2022 [YR2022]',
    '2023 [YR2023]',
    '2024 [YR2024]'
]

for col in years:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

### Cek shape dataset 

```python
 print(df.shape)
```

### Analisis Sparsity

Menghitung persentase missing value pada dataset.

```python
missing = df.isnull().sum().sum()

total = df.size

sparsity = (missing / total) * 100

print("Sparsity:", round(sparsity, 2), "%")
```

# Statistik Deskriptif

Melihat distribusi nilai dataset seperti mean, median, minimum, dan maksimum.

```python
df[years].describe()
```

### Visualisasi Outlier

Mengidentifikasi outlier menggunakan boxplot.

```python
plt.figure(figsize=(10,6))

df[years].boxplot()

plt.title('Boxplot Dataset Kesehatan')

plt.xticks(rotation=45)

plt.show()
```
