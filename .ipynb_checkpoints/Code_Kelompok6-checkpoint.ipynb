## Ubah Missing Value 

```python
df_data = df_data.replace("..", np.nan)
```

```python
#Hitung Missing value
df_data.isnull().sum()
```

```python
# Hitung Sparsity
sparsity = (
    df_data.isnull().sum().sum()
    / (df_data.shape[0] * df_data.shape[1])
) * 100

print("Sparsity:", round(sparsity, 2), "%")
```

```python
# kolom tahun jadi numerik 
years = [
    '2020 [YR2020]',
    '2021 [YR2021]',
    '2022 [YR2022]',
    '2023 [YR2023]',
    '2024 [YR2024]'
]
```

```python
for col in years:
    df_data[col] = pd.to_numeric(df_data[col])
```

```python
# Statistik Deskriptif
df_data[years].describe()
```

```python
# Boxplot
import matplotlib.pyplot as plt

df_data[years].boxplot()

plt.title("Boxplot Data Kesehatan")
plt.show()
```

## Wide Long Transformation

```python
years = [
    '2020 [YR2020]',
    '2021 [YR2021]',
    '2022 [YR2022]',
    '2023 [YR2023]',
    '2024 [YR2024]'
]
```

```python
df_long = pd.melt(
    df_data,
    id_vars=[
        'Series Name',
        'Series Code',
        'Country Name',
        'Country Code'
    ],
    value_vars=years,
    var_name='year',
    value_name='health_value'
)
```

```python
# membersihkan tahun
df_long['year'] = (
    df_long['year']
    .str.extract(r'(\d{4})')
)
```
